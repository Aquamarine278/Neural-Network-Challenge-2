## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head(15)
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
display(attrition_df['Attrition'].value_counts())
display(attrition_df['Department'].value_counts())
```

```python
# let look at the dataset a little more in depth
attrition_df.info()
```

```python
# Create a list of at least 10 column names to use as X data
feature_columns = ['Education', 'Age',  'DistanceFromHome',
                   'EnvironmentSatisfaction', 'HourlyRate',
                   'JobInvolvement', 'JobLevel',  'JobSatisfaction', 'OverTime',
                   'PerformanceRating', 'StockOptionLevel',
                   'WorkLifeBalance','YearsAtCompany', 'YearsSinceLastPromotion'
                   ]


# Create X_df using your selected columns
X_df = attrition_df[feature_columns]


# Show the data types for X_df
print(X_df.dtypes)
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=42)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary

print(X_df['OverTime'].value_counts())
```

```python
# use OneHotEncoder to encode the 'Overtime' column
# drop one of the columns to avoid multicollinearity
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore', drop= 'first')
ohe.set_output(transform="pandas")
overtime_train_enc = ohe.fit_transform(X_train[['OverTime']])
overtime_test_enc = ohe.transform(X_test[['OverTime']])
```

```python
# add the encoded train overtime column to the X_train and drop the original 'Overtime' column
X_train = pd.concat([X_train, overtime_train_enc], axis=1)
X_train.drop(columns=['OverTime'], inplace=True)
# add the encoded test overtime column to the X_test and drop the original 'Overtime' column
X_test = pd.concat([X_test, overtime_test_enc], axis=1)
X_test.drop(columns=['OverTime'], inplace=True)
```

```python
# Create a StandardScaler
scaler =  StandardScaler()

# Fit the StandardScaler to the training data
scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Create a OneHotEncoder for the Department column
dept_ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Fit the encoder to the training data
dept_ohe.fit(y_train[['Department']])

# Create two new variables by applying the encoder
# to the training and testing data
department_train_enc = dept_ohe.transform(y_train[['Department']])
department_test_enc = dept_ohe.transform(y_test[['Department']])

print(department_test_enc)
```

```python
# Create a OneHotEncoder for the Attrition column
att_ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Fit the encoder to the training data
att_ohe.fit(y_train[['Attrition']])
# Create two new variables by applying the encoder
# to the training and testing data
attrition_train_enc = att_ohe.transform(y_train[['Attrition']])
attrition_test_enc = att_ohe.transform(y_test[['Attrition']])

print(attrition_train_enc)
```

## Create, Compile, and Train the Model

```python
X_train_scaled.shape[1]
```

```python
# Find the number of columns in the X training data
print(X_train_scaled.shape[1])

# Create the input layer
input_layer = layers.Input(shape= (X_train_scaled.shape[1],), name= 'input features')

# Create at least two shared layers
shared_layer1 = layers.Dense(units= 64, activation= 'relu', name= 'shared_layer1')(input_layer)
shared_layer2 = layers.Dense(units= 128, activation= 'relu', name= 'shared_layer2')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
department_hidden = layers.Dense(units= 32, activation= 'relu', name= 'department_hidden')(shared_layer2)

# Create the output layer
department_output = layers.Dense(units= 3, activation= 'softmax', name= 'department_output')(department_hidden)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_hidden = layers.Dense(units= 32, activation= 'relu', name= 'attrition_hidden')(shared_layer2)

# Create the output layer
attrition_output = layers.Dense(units= 2, activation= 'sigmoid', name= 'attrition_output')(attrition_hidden)
```

```python
# Create the model
model = Model(inputs= input_layer, outputs= [department_output, attrition_output])

# Compile the model
model.compile(optimizer = 'adam',
              loss = {'department_output' : 'categorical_crossentropy', 'attrition_output': 'binary_crossentropy'},
              metrics = {'department_output' : 'accuracy', 'attrition_output': 'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(X_train_scaled,
          {'department_output': department_train_enc, 'attrition_output': attrition_train_enc},
          epochs= 100,
          batch_size= 32,
          validation_data= (X_test_scaled, {'department_output': department_test_enc, 'attrition_output': attrition_test_enc}))
```

```python
# Evaluate the model with the testing data
results = model.evaluate(X_test_scaled, {'department_output': department_test_enc, 'attrition_output': attrition_test_enc})
print(results)
```

```python
# Print the accuracy for both department and attrition
print(f'Department Accuracy: {results[4]}')
print(f'Attrition Accuracy: {results[3]}')
```
